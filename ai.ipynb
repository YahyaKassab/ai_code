```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import joblib

import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from gensim.models import KeyedVectors
from sklearn.feature_extraction.text import TfidfVectorizer

# Download NLTK resources if not already downloaded
nltk.download('punkt')
nltk.download('stopwords')

# Load FastText embeddings
fasttext_file = 'cc.ar.300.bin'
print("file:", fasttext_file)
try:
    fasttext_model = KeyedVectors.load_word2vec_format(fasttext_file, binary=True, encoding='latin1')
except UnicodeDecodeError as e:
    print(f"UnicodeDecodeError: {e}")

# Load your data
data_path = 'meq_dummy_data.csv'
df = pd.read_csv(data_path)

# Preprocess your data (tokenization, removing stopwords, etc.)
stop_words = set(stopwords.words('arabic'))

def preprocess_text(text):
    tokens = word_tokenize(text)
    filtered_tokens = [token for token in tokens if token not in stop_words]
    return ' '.join(filtered_tokens)

# Combine features into a single input for vectorization
df['question_keywords'] = df.apply(lambda row: preprocess_text(row['question'] + ' ' + row['optimalAnswer'] + ' ' + row['keywords']), axis=1)
df['student_answer_processed'] = df['student_answer'].apply(preprocess_text)

# Vectorize text data using TF-IDF
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df['question_keywords'])
X_student = vectorizer.transform(df['student_answer_processed'])
X_combined = X + X_student  # Combine with student answer features
y = df['score']

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_combined, y, test_size=0.2, random_state=42)

# Train your model
model = LinearRegression()
model.fit(X_train, y_train)

# Evaluate your model
train_score = model.score(X_train, y_train)
test_score = model.score(X_test, y_test)
print(f"Training R^2 Score: {train_score}")
print(f"Test R^2 Score: {test_score}")

# Save model and vectorizer for later use
joblib.dump(model, 'menbar_meq_marking.pkl')
joblib.dump(vectorizer, 'tfidf_vectorizer.pkl')
```

```python
def predict_score(input_data):
    question = input_data.get('question')
    optimalAnswer = input_data.get('optimalAnswer')
    keywords = input_data.get('keywords')
    student_answer = input_data.get('student_answer')

    # Combine question, optimal answer, and keywords
    question_keywords = question + ' ' + optimalAnswer + ' ' + keywords

    # Use FastText embeddings for vector representation
    combined_features = preprocess_text(question_keywords + ' ' + student_answer)
    embedding_vector = sum(fasttext_model[word] for word in combined_features.split() if word in fasttext_model.vocab)

    # Predict the score directly (without vectorization)
    predicted_score = model.predict(embedding_vector.reshape(1, -1))

    return predicted_score[0]

# Example usage with provided Arabic text data
predicted_score_arabic = predict_score({
    "question": "ما هي عاصمة فرنسا؟",
    "optimalAnswer": "عاصمة فرنسا هي باريس.",
    "keywords": "فرنسا عاصمة باريس جغرافيا",
    "student_answer": "باريس هي عاصمة فرنسا."
})

print(f"Predicted Score (Arabic): {predicted_score_arabic}")
```
